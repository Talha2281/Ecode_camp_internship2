```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("/content/diabetes.csv")
```

```python
x=ds.iloc[:,:-1]
y=ds['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
from sklearn.neighbors import KNeighborsClassifier # Corrected class name
knn = KNeighborsClassifier(n_neighbors=20) # Corrected parameter name
knn.fit(x_train, Y_train)
```

```python
knn.score(x_test,Y_test)*100
```

```python
knn.score(x_train,Y_train)*100
```

```python
for i in range(1,30):
  knn1=KNeighborsClassifier(n_neighbors=i)
  knn1.fit(x_train,Y_train)
  print(i,knn1.score(x_train,Y_train)*100,knn1.score(x_test,Y_test))
```
